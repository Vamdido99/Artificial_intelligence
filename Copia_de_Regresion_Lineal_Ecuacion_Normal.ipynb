```
import numpy as np
import matplotlib.pyplot as plt
```

# Iniciemos con  la regresi√≥n lineal univariada

```
x1=np.array([1, 1.5, 2, 2.5, 3, 3.5 , 4, 4.5, 5, 5.5, 6])
x0=np.ones(x1.shape)
y=np.array([0.16,0.28,0.38,0.47,0.60,0.73,0.79,0.87,0.84,0.86,0.96])
X=np.matrix([x0,x1]).T
Y=np.matrix([y]).T
print("shape x0",x0.shape,"shape x1",x1.shape,"shape X",X.shape,"shape Y",Y.shape)
print(X)
print(Y)
```

```
from sklearn.metrics import r2_score
Theta=np.linalg.inv(X.T*X)*(X.T)*Y
print(Theta)
plt.plot(x1,y, 'bo')
plt.plot(x1,Theta[0,0]+Theta[1,0]*x1)
plt.title("Final")
plt.show()
R2 = r2_score(y, Theta[0,0]+Theta[1,0]*x1) #coefficient of determination
print("coefficient of determination = ",R2)
```

**Ejercicio de vacunados en Colombia**

```
#3 Regresiones para total vacunados Colombia
import pandas as pd
df = pd.read_csv("country_vaccinations.csv", sep=';', header=0)
arr = df.values
Xb = arr[:,2:3]
Xa = np.ones((len(Xb),1))
X = np.concatenate((Xa,Xb),axis=1)
Y = arr[:,8] #4
X = np.array(X,dtype='float64') 
Y = np.array(Y,dtype='float64') 
print(Y[1])
```

```
from sklearn.metrics import r2_score
Theta=np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,Y)) 
print(Theta)
plt.plot(Xb,Y)
plt.plot(Xb,Theta[0]+Theta[1]*Xb) #Regresi√≥n Lineal
plt.title("Regresi√≥n Lineal")
plt.xlabel("D√≠as")
plt.ylabel("Total vacunados")
plt.show()

Xx = Theta[0]+Theta[1]*Xb
r2 = r2_score(Y, Theta[0]+Theta[1]*Xb)
print("Coefficient of determination = ",r2)
```

```
X = np.concatenate((Xa,Xb,Xb*Xb),axis=1)
Y = arr[:,8]
X = np.array(X,dtype='float64')
Y = np.array(Y,dtype='float64') 
Theta=np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,Y)) 
print(Theta)
plt.plot(Xb,Y)
plt.plot(Xb,Theta[0]+Theta[1]*Xb+Theta[2]*Xb*Xb) #Regresi√≥n polinomica grado 2
plt.title("Regresi√≥n Polinomial grado 2")
plt.xlabel("D√≠as")
plt.ylabel("Total vacunados")
plt.show()

Xx = Theta[0]+Theta[1]*Xb+Theta[2]*Xb*Xb
r2 = r2_score(Y, Theta[0]+Theta[1]*Xb+Theta[2]*Xb*Xb)
print("Coefficient of determination = ",r2)
```
